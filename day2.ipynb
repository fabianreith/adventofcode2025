## Part II

```python
def check_pattern(snum, pattern):
    len_pattern = len(pattern)
    for i in range(0, len(snum), len_pattern):
        if snum[i:i+len_pattern] != pattern:
            return False
    return True
```

```python
funny_numbers = []
for dpoint in data[1:]:
    num1, num2 = map(int, dpoint.split('-'))
    for num in range(num1, num2 + 1):
        snum = str(num)
        pattern = ''
        for char in snum[:len(snum)//2]:
            pattern += char
            if check_pattern(snum, pattern):
                # print(f"Found pattern: {pattern} in number {snum}")
                funny_numbers.append(num)
                break
        
        
print(np.sum(funny_numbers))
        
```
