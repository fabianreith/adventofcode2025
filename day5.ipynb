```python
with open('day5_input.txt', 'r') as file:
    data = file.read()
```

```python
ranges = data.split('\n\n')[0].splitlines()
ranges = sorted([tuple(map(int, r.replace('-', ' ').split())) for r in ranges], key=lambda x: x[0])
ids = sorted([int(i) for i in data.split('\n\n')[1].splitlines()])
```

## Part I

```python
# Function to check if an ID is in any of the ranges
def is_id_in_ranges(id, ranges):
    for r in ranges:
        if r[0] <= id <= r[1]:
            return True
    return False


# Count how many IDs are in the ranges
count = sum(1 for id in ids if is_id_in_ranges(id, ranges))
print(count)
```

## Part II

```python
ranges = data.split('\n\n')[0].splitlines()
ranges = sorted([tuple(map(int, r.replace('-', ' ').split())) for r in ranges], key=lambda x: x[0])
ids = sorted([int(i) for i in data.split('\n\n')[1].splitlines()])
```

```python
def is_id_in_ranges_list(id, ranges):
    in_ranges = []
    for r in ranges:
        if r[0] <= id <= r[1]:
            in_ranges.append(r)
    return in_ranges
```

```python
validated_ranges = []
for range in ranges:
    r_low = range[0]
    r_high = range[1]
    if is_id_in_ranges_list(r_low, validated_ranges) == [] and is_id_in_ranges_list(r_high, validated_ranges) == []:
        validated_ranges.append(range)
    else:
        for er in set(is_id_in_ranges_list(r_low, validated_ranges) + is_id_in_ranges_list(r_high, validated_ranges)):
            r_low = min(r_low, er[0])
            r_high = max(r_high, er[1])
            validated_ranges.remove(er)
        validated_ranges.append((r_low, r_high))

sum_ranges = 0
for r in validated_ranges:
    sum_ranges += r[1] - r[0] + 1
print(sum_ranges)
```

```python
# O(N) merging approach, suggested by AI

merged_ranges = []

for start, end in ranges:
    # If list is empty or current range starts after the last one ends
    if not merged_ranges or start > merged_ranges[-1][1]:
        merged_ranges.append((start, end))
    else:
        # Overlap detected, update the end of the last range if needed
        merged_ranges[-1] = (merged_ranges[-1][0], max(merged_ranges[-1][1], end))

# Calculate Sum
total_fresh = sum(end - start + 1 for start, end in merged_ranges)
print(total_fresh)
```
