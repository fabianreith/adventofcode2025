# Day 3

```python
import numpy as np
```

```python
with open('day3_input.txt', 'r') as file:
    data = file.read()
data = data.splitlines()
```

## Part I

```python
def get_max_joltage(str_line):
    num_arr = np.array([int(s) for s in str_line])
    first_idx = np.where(num_arr[:-1] == np.max(num_arr[:-1]))[0][0]
    second_idx = np.where(num_arr[first_idx+1:] == np.max(num_arr[first_idx+1:]))[0][0]
    final_num = str(num_arr[first_idx]) + str(num_arr[first_idx + 1 + second_idx])
    return int(final_num)
```

```python
max_jolt_list = []
for line in data:
    max_jolt = get_max_joltage(line)
    max_jolt_list.append(max_jolt)
print(sum(max_jolt_list))
```

## Part II

```python
def get_max_joltage_gen(str_line, num=2):
    num_arr = np.array([int(s) for s in str_line])
    idx_arr = []
    idx_left = 0
    for idx_right in range(len(num_arr)-num+1, len(num_arr)+1):
        i_idx = np.where(num_arr[idx_left:idx_right] == np.max(num_arr[idx_left:idx_right]))[0][0]
        idx_arr.append(i_idx + idx_left)
        idx_left = i_idx + idx_left + 1
    final_num = ''.join([str(num_arr[i]) for i in idx_arr])
    return int(final_num)
```

```python
max_jolt_list = []
for line in data:
    max_jolt = get_max_joltage_gen(line, num=12)
    max_jolt_list.append(max_jolt)
print(sum(max_jolt_list))
```
