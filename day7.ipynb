## Part II

```python
# This is a recursive function that explores all beam paths - but it takes too long to work out :(
def get_beam_timelines(lines, idxs_S=[], idxs_I=[]):
    for i, line in enumerate(lines):
        if i == len(lines)-1:
            return 1 # should be 1
        for idx in set(idxs_S + idxs_I):
            symbol = line[idx]
            if symbol == '^':
                line_left = line.copy()
                line_right = line.copy()
                line_left[idx-1] = '|'
                line_right[idx+1] = '|'
                idxs_I_left = [i for i, char in enumerate(line_left) if char == '|']
                idxs_I_right = [i for i, char in enumerate(line_right) if char == '|']
                time_lines = get_beam_timelines(lines[i+1:], idxs_I=idxs_I_left) + get_beam_timelines(lines[i+1:], idxs_I=idxs_I_right)
                return time_lines
            else:
                line[idx] = '|'
        # get all the indices where string test has an 'S' char in it'
        idxs_S = [i for i, char in enumerate(line) if char == 'S']
        idxs_I = [i for i, char in enumerate(line) if char == '|']


time_lines = get_beam_timelines(lines[:10])
print(time_lines)
```

```python
with open('day6_input.txt', 'r') as file:
    data = file.read()
data = data.splitlines()
data = [list(line) for line in data]
data = np.array(data)
cep_data = data[:, ::-1]
```

```python
print(data[0][::-1][:20])
print(data[1][::-1][:20])
print(data[2][::-1][:20])
print(data[3][::-1][:20])
print(data[4][::-1][:20])
```

```python
cep_sum = 0
nums = []
for col in cep_data.T:
    num = ''
    for n in col[:-1]:
        if n != ' ':
            num += n
    
    try:
        nums.append(int(num))
    except:
        # not a number line
        continue
    operator = col[-1]
    if operator != ' ':
        # last number line here
        if operator == "+":
            cep_sum += np.sum(nums)
        elif operator == "*":
            cep_sum += np.prod(nums)
        nums = []
            
        
print(cep_sum)  
```

```python
# Proposed AI solution I prompted after solving the problem myself

import math

# 1. Read and pad lines to ensure the grid is rectangular
with open('day6_input.txt') as f:
    lines = f.read().splitlines()

# 2. Transpose (swap rows/cols) using zip
#    This creates a list of columns (tuples of characters)
#    We use reversed() to handle the "Right-to-Left" requirement
cols = list(zip(*lines))

total = 0
current_nums = []

for col in reversed(cols):
    # 3. Clean Logic: Skip empty columns immediately
    if all(c == ' ' for c in col):
        continue

    # 4. Extract number: Join digits vertically
    num_str = "".join(c for c in col if c.isdigit())
    if num_str:
        current_nums.append(int(num_str))

    # 5. Check Operator: The prompt says it's at the bottom
    operator = col[-1]

    if operator == '+':
        total += sum(current_nums)
        current_nums = []  # Reset for next problem
    elif operator == '*':
        total += math.prod(current_nums)
        current_nums = []

print(total)
```
