## Part I

```python
with open('day7_input.txt', 'r') as file:
    data = file.read()
lines = data.splitlines()
lines = [list(line) for line in lines]
```

```python
idxs_S = []
idxs_I = []
beam_splits = 0
for line in lines:
    for idx in set(idxs_S + idxs_I):
        symbol = line[idx]
        if symbol == '^':
            line[idx-1] = '|'
            line[idx+1] = '|'
            beam_splits += 1
        else:
            line[idx] = '|'
    # get all the indices where string test has an 'S' char in it'
    idxs_S = [i for i, char in enumerate(line) if char == 'S']
    idxs_I = [i for i, char in enumerate(line) if char == '|']    
print(beam_splits)
```

## Part II

```python
import copy
```

```python
# Define your mapping
mapping = {'.': 0, 'S': -1, '^': 2}

lines = []
with open('day7_input.txt', 'r') as file:
    for line in file:
        # Get each char, look it up in mapping, default to int(char) if not found
        row = [mapping.get(char, char) for char in line.strip()]
        lines.append([int(x) for x in row])

np_lines = np.array(lines)
np_lines_backup = np_lines.copy()
lines_backup = copy.deepcopy(lines)
```

```python
idxs_1 = []
idxs_2 = []
lines = copy.deepcopy(lines_backup)
for i in range(len(lines)):
    line = lines[i]
    for idx in set(idxs_1 + idxs_2):
        symbol = line[idx]
        if symbol == 2:
            line[idx-1] += lines[i-1][idx]
            line[idx+1] += lines[i-1][idx]
        elif symbol <= 0:
            line[idx] += lines[i-1][idx]
    idxs_1 = [i for i , num in enumerate(line) if num < 0]
    # idxs_2 = [i for i, num in enumerate(line) if num == 2]
    lines[i] = line
    # print(line)
print("The result is:", -np.sum(lines[-1]))
```
