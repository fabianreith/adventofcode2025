# Day 1

```python
with open('day1_input.txt', 'r') as file:
    data = file.read()
data = data.split("\n")
print(data[0])
print(data[-1])
```

## Part I

```python
current_point = 50
visited_zero = 0
for instr in data:
    if instr.startswith('L'):
        move_command = - int(instr[1:])
    elif instr.startswith('R'):
        move_command = int(instr[1:])
    current_point = (current_point + move_command) % 100
    if current_point == 0:
        visited_zero += 1
print(visited_zero)
```

## Part II

```python
current_point = 50
visited_zero = 0
for instr in data:
    if instr.startswith('L'):
        move_command = - int(instr[1:])
    elif instr.startswith('R'):
        move_command = int(instr[1:])
    to_move_point = (current_point + move_command)
    if current_point == 0 and to_move_point < 0:
        visited_zero -= 1 # if current point is 0 - won't visit zero again
    while to_move_point > 100:
        to_move_point -= 100
        visited_zero += 1
    while to_move_point < 0: 
        to_move_point += 100
        visited_zero += 1 
    if to_move_point == 100:
        pass
    current_point = to_move_point % 100
    if current_point == 0:
        visited_zero += 1
print(visited_zero)
```

```python
current_absolute_pos = 50
total_zeros = 0

# Parse instructions
# We assume 'data' is the list of strings like ["L68", "R48"]
for instr in data:
    direction = instr[0]
    amount = int(instr[1:])
    
    start = current_absolute_pos
    
    if direction == 'R':
        end = start + amount
        # Count multiples of 100 in (start, end]
        # Formula: floor(end/100) - floor(start/100)
        crossings = (end // 100) - (start // 100)
        total_zeros += crossings
        current_absolute_pos = end
        
    elif direction == 'L':
        end = start - amount
        # Count multiples of 100 in [end, start)
        # Formula: floor((start-1)/100) - floor((end-1)/100)
        # We subtract 1 to handle the exclusive/inclusive boundary correctly
        crossings = ((start - 1) // 100) - ((end - 1) // 100)
        total_zeros += crossings
        current_absolute_pos = end
print(total_zeros)
```
